```python
import pandas as pd
```

```python
df=pd.DataFrame(
    {

    'City': [
        'Mumbai', 'Pune', 'Delhi', 'Bangalore', 'Hyderabad',
        'Chennai', 'Kolkata', 'Jaipur', 'Ahmedabad', 'Surat',
        'Nagpur', 'Indore', 'Lucknow', 'Patna', 'Bhopal',
        'Vadodara', 'Coimbatore', 'Chandigarh', 'Nashik', 'Rajkot'
    ],
    'Price': [
        9500000, 7200000, 8800000, 9300000, 8500000,
        8100000, 7800000, 6700000, 7400000, 7100000,
        6600000, 6900000, 7300000, 6400000, 7000000,
        6800000, 7500000, 7700000, 7000000, 6900000
    ]
}
    
)
```

```python
df
```

```python
meain_price=df.groupby('City')['Price'].mean().to_dict()
```

```python
meain_price
```

```python
df['city_encoded']=df['City'].map(meain_price)
```
