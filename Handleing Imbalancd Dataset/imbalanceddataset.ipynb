```python
import numpy as nd
import pandas as pd

#set the random seed for reproducibility

nd.random.seed(123)

#create a dataframe with two classes
n_sample=1000
class_0_reatio=0.9
n_class_0=int(n_sample *class_0_reatio)
n_class_1=n_sample-n_class_0
```

```python
n_class_0,n_class_1
```

```python
class_0=pd.DataFrame({
    
    'Name': ['Amit', 'Sneha', 'Rohit', 'Priya', 'Kunal'],
    'Age': [23, 21, 24, 22, 23],
    'City': ['Pune', 'Mumbai', 'Nagpur', 'Nashik', 'Kolhapur'],
    'Marks': [85, 90, 78, 88, 92]

})

class_1=pd.DataFrame({
    'Experience_Years': [1, 3, 5, 7, 2, 4, 6, 8],
    'Age': [22, 25, 28, 32, 24, 27, 30, 35],
    'Education_Level': ['Bachelors', 'Masters', 'Bachelors', 'PhD', 'Bachelors', 'Masters', 'Masters', 'PhD'],
    'Salary': [25000, 40000, 55000, 80000, 30000, 50000, 65000, 90000],
    'Promoted': [0, 0, 1, 1, 0, 1, 1, 1]  # Target column (0 = Not promoted, 1 = Promoted)
})
```

```python
df=pd.concat([class_0,class_1]).reset_index(drop=True)
```
